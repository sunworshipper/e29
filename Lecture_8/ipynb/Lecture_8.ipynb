## Accuracy takehomes
<p class="grey"> Better metrics are required for evaluating models </p>
- - -
* Majority and Random can perform as well as Learned
* Data is often imbalanced, you will definitely run into this kind of data
* You should not use accuracy as a metric in an imbalanced setting
* We need better metrics that can tell us how much better a model performs than most frequest or random predictions

## Confusion Matrix
<p class="grey"> A comprehensive method for evaluating classification models </p>
- - -
* It's an NXN matrix where N is the number of classes (labels)
* What kind of matrix would you have in binary classification?
* It shows us the number of correct and incorrect predictions made by our model compared to the actual labels
* We can use the data in the matrix to evaluate model performance by calculating several parameters

## PIMA INDIANS DIABETES DATASET
<p class="grey"> KNN Example  </p>
- - -

```python
# Load CSV using Pandas from URL
#DESCRIPTION: http://mlearn.ics.uci.edu/databases/pima-indians-diabetes/pima-indians-diabetes.names
url = 'http://mlearn.ics.uci.edu/databases/pima-indians-diabetes/pima-indians-diabetes.data'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
pima = pd.read_csv(url, names=names)
print(pima.shape)

# Create arrays for the features and the response variable
y = pima['class'].values
X = pima.drop('class', axis=1).values

#Import KNN Classifier
from sklearn.neighbors import KNeighborsClassifier

# Import classification_report and confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

# Split Data to Test And Train data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.4,random_state=42)

# Instantiate KNN classifier
knn = KNeighborsClassifier(n_neighbors=6)

# Fit the classifier to the training data
knn.fit(X_train,y_train)

# Predict the labels of the test data: y_pred
y_pred = knn.predict(X_test)

# Generate the confusion matrix and classification report
print ("THIS IS THE CONFUSION MATRIX")
print(confusion_matrix(y_test, y_pred))
print ("THIS IS THE CLASSIFICATION REPORT")
print(classification_report(y_test, y_pred))

print ("Where 0 is predicted to not get Diabetes, and 1 means predicted to test positive for Diabetes")
print(knn.score(X_test, y_test))
```

```python
#Code adopted from Jason Brownlee
#Jason has many free tutorials and sells Python ML books
#This is not a bad way to evaluate multiple classifiers on a same dataset if your data was balanced
#You have to do a bit more work to tune classifiers independantly 
#This is just so you can get the idea that you should try different things.

from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
# prepare configuration for cross validation test harness
seed = 7
# prepare models
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X, y, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
    
    
# boxplot algorithm comparison
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston

boston = load_boston()
X = boston.data
y = boston.target

#we split the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42)

reg = LinearRegression()
reg.fit(X_train,y_train)

prediction = reg.predict(X_test)
prediction

#R SQUARED
rsq = reg.score(X_test,y_test)
print ("R^2:",rsq )

#ROOT MEAN SQUARED ERROR
from sklearn.metrics import mean_squared_error
#square root of the variance of the residuals
#a measure of how close the observed data points are to the models predicted values
rmse = np.sqrt(mean_squared_error(y_test,prediction))
print ("RMSE:",rmse)
```

## Regularized regression with Lasso and why it's awesome!
<p class="grey"> Ever wondered how to pick your features?  </p>
- - -

* Lasso has a tendency to minimize the coefficients of less contributing features to 0
* The Lasso algorithm will use the non-zero ones for the model

\begin{equation*}
Lasso  J(w) = OLS + \lambda * \sum_{i=1}^n |w_i|
\end{equation*}

```python
#Note that we are still working with the Boston Housing Dataset

from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1, normalize = True)

# Fit the model
lasso.fit(X,y)

#Calculate the coefficients
lasso_coef = lasso.coef_ 
print("These are the calculated coefficients:", lasso_coef)

# Plot the coefficients
g = plt.plot(range(len(boston.feature_names)), lasso_coef)
g = plt.xticks(range(len(boston.feature_names)), boston.feature_names, rotation=60)
g = plt.ylabel('Coefficients')
plt.show()
```

## Regularized regression Ridge
<p class="grey"> Ever wondered how to pick your features?  </p>
- - -

* Unlike Lasso which minimizes the coefficients to 0 Ridge minimizes them to close to 0 but not 0

\begin{equation*}
Ridge  J(w) = OLS + \lambda * \sum_{i=1}^n (w_i)^2
\end{equation*}

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.1, normalize = True)
ridge.fit (X_train, y_train)
pred = ridge.predict(X_test)
ridge.score(X_test,y_test)
```
